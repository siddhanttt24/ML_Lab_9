```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
digits = load_digits()
X = digits.data  # Features (pixel values)
y = digits.target  # Target labels (digits 0-9)

print("Dataset Shape:", X.shape)
print("Number of Classes:", len(np.unique(y)))
```

```python
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

```python
pca = PCA()
pca.fit(X_standardized)

# Plot the explained variance
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Components')
plt.grid()
plt.show()
```

```python
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_standardized)
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='tab10', alpha=0.7)
plt.title('PCA of Digits Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, ticks=range(10), label='Digits')
plt.grid()
plt.show()
```

```python
print("Explained Variance Ratio by Principal Component:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"Component {i+1}: {var:.4f}")
```
